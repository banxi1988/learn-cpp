```python
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
import csv
x_list = []
y_list = []
with open("malloc.csv") as csvfile:
    reader = csv.reader(csvfile)
    for (size,gap) in reader:
        if not (size and gap):
            continue
        size = int(size)
        gap = int(gap)
        if gap < 0 or gap > (size * 4):
            continue
        x_list.append(size)
        y_list.append(gap)

fig = plt.figure(figsize=(10.24,7.68))
ax = fig.subplots()
ax.set_xlabel('malloc size')
ax.set_ylabel('pointer gap')
ax.set_title('malloc pointer gap')
ax.grid(which='major', axis='y', c='#ccc')
locator = ticker.LogLocator(2)
ax.yaxis.set_major_locator(locator)
ax.xaxis.set_major_locator(locator)
ax.scatter(x_list,y_list, marker='.', label='Gap')
fig.tight_layout()
fig.show()
fig.savefig("malloc_32Kx4.png")


```
